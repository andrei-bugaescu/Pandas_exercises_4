- Which actor is mentioned the most often?


```
import pandas as pd

link = 'https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/imdb_1000.csv'

df_actor = pd.read_csv(link,sep=',')
df_actor.head()
```

```
df_actor['actors_list'] = df_actor['actors_list'].apply(eval)
df_actor.head()
```

```
df_actor = df_actor.explode('actors_list')
df_actor
```

```
df_actor['actors_list'].value_counts().idxmax()
```
